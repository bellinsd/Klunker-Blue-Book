```python
from __future__ import print_function, division

%matplotlib inline

import numpy as np
import pandas as pd
import thinkstats2
import thinkplot
import matplotlib.pyplot as plt
import random
```

```python
from thinkstats2 import Mean, MeanVar, Var, Std, Cov

def LeastSquares(xs, ys):
    meanx, varx = MeanVar(xs)
    meany = Mean(ys)

    slope = Cov(xs, ys, meanx, meany) / varx
    inter = meany - slope * meanx

    return inter, slope
```

```python
def FitLine(xs, inter, slope):
    fit_xs = np.sort(xs)
    fit_ys = inter + slope * fit_xs
    return fit_xs, fit_ys
```

```python
#AGE AND PRICE scatterplot with trend line
inter, slope = LeastSquares(df["AGE"], df["Price Fs"])
inter, slope
fit_xs, fit_ys = FitLine(df["AGE"], inter, slope)
thinkplot.Scatter(df["AGE"], df["Price Fs"], color='blue', alpha=0.1, s=10)
thinkplot.Plot(fit_xs, fit_ys, color='white', linewidth=3)
thinkplot.Plot(fit_xs, fit_ys, color='red', linewidth=2)
thinkplot.Config(xlabel="Age of Car",
                 ylabel='Price',
                 axis=[0, 30, 0, 100000],
                 legend=False)
```
