```python
import pandas as pd
```

```python
#Reading in CSV files
car_data_clean = pd.read_csv('car_data_clean.csv')
```

```python
#Checking intitial car dataset
car_data_clean.describe()
```

```python
car_data_clean.isnull().sum()
```

```python
car_data_clean = car_data_clean.dropna()
```

```python
car_data_clean.isnull().sum()
```

```python
car_data_clean.describe()
```

```python
#Reading in the other datasets
county_demo = pd.read_csv('county_demos_cleaned.csv')
zip_county = pd.read_csv('zip_code_database_cleaned.csv')
age = pd.read_csv('year_age.csv')
region = pd.read_csv('regions_states.csv')
```

```python
#Inner Join: County Demographic and Zip Datasets
df_countydemo_zip = pd.merge(zip_county, county_demo, how='inner', left_on=['county','state'], right_on=['County','State short'])
```

```python
#Checking merge of previous cell
#df_countydemo_zip.tail
df_countydemo_zip.describe()
```

```python
df_countydemo_zip.isnull().sum()
```

```python
df_countydemo_zip.to_csv('countydemo_zip.csv', index=False, header=True)
```

```python
#Inner Join: Original Car Data and Age Datasets
df_car_age = pd.merge(car_data_clean, age, how='inner', left_on='year_is', right_on='YEAR')
df_car_age.head
```

```python
df_car_age.isnull().sum()
```

```python
#df_car_age.tail
df_car_age.describe()
```

```python
df_car_age.to_csv('car_age.csv', index=False, header=True)
```

```python
#Inner Join: Countydemo_zip and Region Datasets
df_county_zip_region = pd.merge(df_countydemo_zip, region, how='inner', left_on='State short', right_on='States Short')
df_county_zip_region.head
```

```python
#Checking merge of previous cell
df_county_zip_region.describe()
#df_county_zip_region.isnull().sum()
```

```python
df_county_zip_region.to_csv('county_zip_region.csv', index=False, header=True)
```

```python
#Changing type in 'zip' column (type)
#df_car_age['zip_is'] = df_car_age['zip_is'].astype()
df_county_zip_region['zip'] = df_county_zip_region['zip'].astype(str)
```

```python
#Inner Join: Final
df = pd.merge(df_car_age, df_county_zip_region, how='inner', left_on='zip_is', right_on='zip')
df.head
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
#Save
df.to_csv('car_data_clean_complete_v3.csv', index=False, header=True)
```
